# Distribucin Uniforme U(0,1)

## a. Funcin de Densidad y Distribucin Acumulada

### Funcin de Densidad \(f(x)\)

La funcin de densidad de una distribucin uniforme \(U(a, b)\) se define como:

$$
f(x) = 
\begin{cases} 
\frac{1}{b - a} & \text{para } a \le x \le b \\
0 & \text{de lo contrario}
\end{cases}
$$

Para una distribucin uniforme \(U(0,1)\), los valores de \(a = 0\) y \(b = 1\). Por lo tanto, la funcin de densidad simplificada es:

$$
f(x) = 
\begin{cases} 
1 & \text{para } 0 \le x \le 1 \\
0 & \text{de lo contrario}
\end{cases}
$$

### Funcin de Distribucin Acumulada \(F(x)\)

La funcin de distribucin acumulada se define como la integral de la funcin de densidad desde el lmite inferior de la distribucin hasta \(x\):

$$
F(x) = \int_{a}^{x} f(t) \, dt
$$

Para la distribucin uniforme \(U(0,1)\), esto se convierte en:

$$
F(x) = 
\begin{cases} 
0 & \text{para } x < 0 \\
x & \text{para } 0 \le x \le 1 \\
1 & \text{para } x > 1
\end{cases}
$$


## Pasos de Integracin de la Funcin Acumulada para U(a,b)

Para encontrar la funcin de distribucin acumulada \( F(x) \) de una distribucin uniforme general \( U(a,b) \), realizamos la integracin de su funcin de densidad \( f(x) \). Recordemos que la funcin de densidad para U(a,b) es:

$$
f(x) = 
\begin{cases} 
\frac{1}{b - a} & \text{para } a \le x \le b \\
0 & \text{de lo contrario}
\end{cases}
$$

### Integracin de la funcin de densidad

La funcin de distribucin acumulada \( F(x) \) se define como:

$$
F(x) = \int_{-\infty}^{x} f(t) \, dt
$$

Para calcular \( F(x) \), consideramos tres casos basados en el valor de \( x \):

1. **Para \( x < a \):**

   Dado que \( f(t) = 0 \) para \( t < a \), la integral es:

   $$
   F(x) = \int_{-\infty}^{x} 0 \, dt = 0
   $$

2. **Para \( a $\leq$ x $\leq$ b \):**

   Dado que \( f(t) = $\frac{1}{b-a}$ \) para \( a $\leq$ t $\leq$ b \), la integral es:

   $$
   F(x) = \int_{a}^{x} \frac{1}{b - a} \, dt = \frac{x - a}{b - a}
   $$

3. **Para \( x > b \):**

   Aqu, la integral se calcula desde \( a \) hasta \( b \) (ya que  $f(t) = \frac{1}{b-a}$  en este intervalo) y es 0 fuera de este intervalo:

   $$
   F(x) = \int_{a}^{b} \frac{1}{b - a} \, dt = \frac{b - a}{b - a} = 1
   $$

Por lo tanto, la funcin de distribucin acumulada \( F(x) \) para la distribucin uniforme U(a,b) se resume como:

$$
F(x) = 
\begin{cases} 
0 & \text{para } x < a \\
\frac{x - a}{b - a} & \text{para } a \le x \le b \\
1 & \text{para } x > b
\end{cases}
$$


## Pasos de Integracin de la Funcin Acumulada para U(0,1)

Para encontrar la funcin de distribucin acumulada \( F(x) \) de una distribucin uniforme \( U(0,1) \), realizamos la integracin de su funcin de densidad \( f(x) \). Recordemos que la funcin de densidad para U(0,1) es:

$$
f(x) = 
\begin{cases} 
1 & \text{para } 0 \le x \le 1 \\
0 & \text{de lo contrario}
\end{cases}
$$

### Integracin de la funcin de densidad

La funcin de distribucin acumulada \( F(x) \) se define como:

$$
F(x) = \int_{-\infty}^{x} f(t) \, dt
$$

Para calcular \( F(x) \), consideramos tres casos basados en el valor de \( x \):

1. **Para \( x < 0 \):**

   Dado que \( f(t) = 0 \) para \( t < 0 \), la integral es:

   $$
   F(x) = \int_{-\infty}^{x} 0 \, dt = 0
   $$

2. **Para \( 0 $\leq$ x $\leq$ 1 \):**

   Dado que \( f(t) = 1 \) para \( 0 $\leq$ t $\leq$ 1 \), la integral es:

   $$
   F(x) = \int_{0}^{x} 1 \, dt = x - 0 = x
   $$

3. **Para \( x > 1 \):**

   Aqu, la integral se calcula desde 0 hasta 1 (ya que \( f(t) = 1 \) en este intervalo) y es 0 fuera de este intervalo:

   $$
   F(x) = \int_{0}^{1} 1 \, dt = 1 - 0 = 1
   $$

Por lo tanto, la funcin de distribucin acumulada \( F(x) \) para la distribucin uniforme U(0,1) se resume como:

$$
F(x) = 
\begin{cases} 
0 & \text{para } x < 0 \\
x & \text{para } 0 \le x \le 1 \\
1 & \text{para } x > 1
\end{cases}
$$


```python
# Definicin del rango de x para graficar
x = np.linspace(-0.1, 1.1, 1000)

# Funcin de densidad para U(0,1)
# f(x) = 1/(b-a) para a <= x <= b, y 0 en otro caso.
# Para U(0,1), esto se simplifica a f(x) = 1 para 0 <= x <= 1.
f_x = np.where((0 <= x) & (x <= 1), 1, 0)

# Funcin de distribucin acumulada para U(0,1)
# F(x) = (x-a)/(b-a) para a <= x <= b, y 0 para x < a, 1 para x > b.
# Para U(0,1), esto se simplifica a F(x) = x para 0 <= x <= 1.
F_x = np.where(x < 0, 0, np.where(x <= 1, x, 1))

# Grficas de las funciones
plt.figure(figsize=[15, 5])

# Grfica de la funcin de densidad f(x)
plt.subplot(1, 2, 1)
plt.plot(x, f_x, 'b-', label='f(x)')
plt.title('Funcin de Densidad U(0,1)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.ylim(0, 1.5)
plt.fill_between(x, f_x, color='blue', alpha=0.3)  # rea bajo la curva

# Grfica de la funcin de distribucin acumulada F(x)
plt.subplot(1, 2, 2)
plt.plot(x, F_x, 'r-', label='F(x)')
plt.title('Funcin de Distribucin Acumulada U(0,1)')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True)
plt.ylim(-0.1, 1.1)

plt.tight_layout()  # Ajusta automticamente los espacios de las subgrficas
plt.show()
```

## Clculo de la Media y Varianza para U(a,b)

La media (esperanza) y la varianza de una distribucin uniforme nos informan sobre el comportamiento central y la dispersin de la distribucin, respectivamente.

### Clculo de la Media (Esperanza) \( E[X] \)

La media o esperanza de una variable aleatoria continua \( X \) con una funcin de densidad \( f(x) \) sobre un intervalo \([a, b]\) se calcula como:

$$
E[X] = \int_{a}^{b} x f(x) \, dx
$$

Para una distribucin uniforme \( U(a, b) \), donde $$ f(x) = \frac{1}{b - a}  $$ para \( a $\le$ x $\le$ b \), la esperanza se calcula como:

$$
E[X] = \int_{a}^{b} x \left(\frac{1}{b - a}\right) dx = \frac{1}{b - a} \left[ \frac{x^2}{2} \right]_{a}^{b} = \frac{1}{b - a} \left( \frac{b^2 - a^2}{2} \right) = \frac{a + b}{2}
$$

### Clculo de la Varianza \( Var(X) \)

La varianza de una variable aleatoria continua se calcula como la esperanza del cuadrado de la desviacin de \( X \) de su media, es decir:

$$
Var(X) = E[X^2] - (E[X])^2
$$

Primero calculamos \( E[$X^2$] \):

$$
E[X^2] = \int_{a}^{b} x^2 \left(\frac{1}{b - a}\right) dx = \frac{1}{b - a} \left[ \frac{x^3}{3} \right]_{a}^{b} = \frac{1}{b - a} \left( \frac{b^3 - a^3}{3} \right)
$$

Luego, la varianza \( Var(X) \) para \( U(a, b) \) es:

$$
Var(X) = \frac{1}{b - a} \left( \frac{b^3 - a^3}{3} \right) - \left(\frac{a + b}{2}\right)^2 = \frac{(b - a)^2}{12}
$$

### Caso Especfico para U(0, 1)

Para la distribucin uniforme \( U(0, 1) \), donde \( a = 0 \) y \( b = 1 \):

1. **Media (Esperanza) \( E[X] \):**

   $$
   E[X] = \frac{0 + 1}{2} = 0.5
   $$

2. **Varianza \( Var(X) \):**

   $$
   Var(X) = \frac{(1 - 0)^2}{12} = \frac{1}{12}
   $$


## Desarrollo Algebraico de la Varianza para U(a,b)

Para calcular la varianza \( Var(X) \) de una distribucin uniforme \( U(a, b) \), utilizamos la frmula:

$$
Var(X) = E[X^2] - (E[X])^2
$$

Donde:

1. \( $E[X^2]$ \) es la esperanza del cuadrado de la variable aleatoria, que se calcula como:

   $$
   E[X^2] = \int_{a}^{b} x^2 \left(\frac{1}{b - a}\right) dx = \frac{1}{b - a} \left[ \frac{x^3}{3} \right]_{a}^{b} = \frac{1}{b - a} \left( \frac{b^3 - a^3}{3} \right)
   $$

   Aqu, usamos la frmula para el clculo de la integral de \( $x^2$ \), que da como resultado \( $\frac{x^3}{3}$ \) evaluado entre \( a \) y \( b \).

2. \( ($E[X])^2$ \) es el cuadrado de la media (esperanza) de la distribucin, que ya hemos calculado que es \( $\frac{a + b}{2}$ \):

   $$
   (E[X])^2 = \left(\frac{a + b}{2}\right)^2 = \frac{a^2 + 2ab + b^2}{4}
   $$

Ahora, sustituimos \( $E[X^2]$ \) y \( ($E[X])^2$ \) en la frmula de la varianza y desarrollamos:

$$
Var(X) = \frac{1}{b - a} \left( \frac{b^3 - a^3}{3} \right) - \frac{a^2 + 2ab + b^2}{4}
$$

Primero, desarrollemos la diferencia de cubos \( $b^3 - a^3$ \):

$$
b^3 - a^3 = (b - a)(b^2 + ba + a^2)
$$

Ahora, insertamos esto en la expresin para \( $E[X^2]$ \):

$$
E[X^2] = \frac{1}{b - a} \left( \frac{(b - a)(b^2 + ba + a^2)}{3} \right) = \frac{b^2 + ba + a^2}{3}
$$

Entonces, la varianza se convierte en:

$$
Var(X) = \frac{b^2 + ba + a^2}{3} - \frac{a^2 + 2ab + b^2}{4}
$$

Para combinar estos trminos en una nica fraccin, encontramos un denominador comn (12) y combinamos:

$$
Var(X) = \frac{4(b^2 + ba + a^2) - 3(a^2 + 2ab + b^2)}{12} = \frac{4b^2 + 4ba + 4a^2 - 3a^2 - 6ab - 3b^2}{12}
$$

Simplificamos los trminos:

$$
Var(X) = \frac{b^2 - 2ba + a^2}{12} = \frac{(b - a)^2}{12}
$$

Este es el desarrollo algebraico que muestra cmo se calcula la varianza de una distribucin uniforme \( U(a, b) \).


# Procedimiento para Realizar la Prueba de Bondad de Ajuste  (Chi-cuadrado)

## a) Realizar la Prueba de Bondad de Ajuste 

La Prueba de Bondad de Ajuste  se utiliza para determinar si una muestra de datos se ajusta a una distribucin esperada. Nos ayuda a comparar si las frecuencias observadas en los datos se ajustan a las frecuencias esperadas de una determinada distribucin terica.

### Pasos:

1. **Formular las Hiptesis**:
   - **H**: No hay diferencia entre las frecuencias observadas y las esperadas (los datos siguen la distribucin esperada).
   - **H**: Existe una diferencia significativa entre las frecuencias observadas y las esperadas (los datos no siguen la distribucin esperada).

2. **Determinar las Categoras y Frecuencias**:
   - Dividir los datos en categoras.
   - Contar cuntas observaciones hay en cada categora (frecuencias observadas).
   - Calcular las frecuencias esperadas para cada categora basadas en la distribucin terica.

3. **Calcular el Estadstico de Prueba**:
   - Utilizar la frmula:  = [(O - E) / E], donde 'O' son las frecuencias observadas y 'E' las esperadas.
   - Sumar los valores para todas las categoras.

4. **Determinar el Valor Crtico o p-valor**:
   - Seleccionar un nivel de significancia (), comnmente 0.05 o 0.01.
   - Encontrar el valor crtico correspondiente usando los grados de libertad (nmero de categoras - 1) en una tabla  o calcular el p-valor.

5. **Tomar una Decisin**:
   - Si el estadstico  es mayor que el valor crtico o si el p-valor es menor que , rechazar H.
   - De lo contrario, no rechazar H.

## b) Leer una tabla estadstica 

1. **Identificar los Grados de Libertad (df)**: Nmero de categoras menos uno.
2. **Localizar el Nivel de Significancia ()**: Umbral para rechazar la hiptesis nula, tpicamente 0.05 o 0.01.
3. **Encontrar el Valor Crtico**: Usar df y  para encontrar el valor en la tabla .
4. **Comparacin**: Si  calculado > valor crtico de la tabla, rechazar H.

## Limitaciones de la Prueba de Bondad de Ajuste 

- **Tamao de la Muestra**: Puede no ser adecuada para muestras muy pequeas.
- **Datos Categricos**: Solo se aplica a datos categricos, no a datos continuos sin categorizacin previa.
- **Esperanzas Bajas**: No debe usarse si el nmero esperado de observaciones en alguna categora es menor a 5.
- **Independencia**: Los datos observados deben ser independientes entre s.


# Ejemplo Detallado de la Prueba de Bondad de Ajuste 

En este ejemplo, consideraremos un conjunto de datos ficticios para entender cmo realizar y interpretar una Prueba de Bondad de Ajuste . Supongamos que estamos investigando si una dado es justo, es decir, si todas las caras tienen la misma probabilidad de aparecer.

## Paso 1: Formular las Hiptesis

- **H (Hiptesis Nula)**: El dado es justo. Las frecuencias observadas se ajustan a las frecuencias esperadas basadas en la distribucin uniforme.
- **H (Hiptesis Alternativa)**: El dado no es justo. Las frecuencias observadas no se ajustan a las frecuencias esperadas.

## Paso 2: Determinar las Categoras y Frecuencias

En este ejemplo, las categoras son los posibles resultados de lanzar un dado, es decir, 1, 2, 3, 4, 5 y 6.

Supongamos que hemos lanzado el dado 60 veces (nuestra muestra), y hemos registrado las siguientes frecuencias observadas para cada resultado:

| Resultado | Frecuencia Observada |
|-----------|----------------------|
| 1         | 11                   |
| 2         | 9                    |
| 3         | 10                   |
| 4         | 10                   |
| 5         | 10                   |
| 6         | 10                   |

Dado que esperamos que cada cara tenga la misma probabilidad de aparecer en un dado justo, las frecuencias esperadas seran todas iguales (la muestra total, 60, dividida entre el nmero de categoras, 6):

| Resultado | Frecuencia Esperada |
|-----------|---------------------|
| 1         | 10                  |
| 2         | 10                  |
| 3         | 10                  |
| 4         | 10                  |
| 5         | 10                  |
| 6         | 10                  |

## Paso 3: Calcular el Estadstico de Prueba

La frmula para calcular el valor de  es:

$ 
\sum \left[ \frac{(O - E)^2}{E} \right]
$

donde '' significa la suma a travs de todas las categoras, 'O' representa la frecuencia observada y 'E' la frecuencia esperada. Calculamos esto para cada categora:

| Resultado | O  | E  | (O-E) | (O-E) / E |
|-----------|----|----|-------|------------|
| 1         | 11 | 10 | 1     | 0.1        |
| 2         | 9  | 10 | 1     | 0.1        |
| 3         | 10 | 10 | 0     | 0          |
| 4         | 10 | 10 | 0     | 0          |
| 5         | 10 | 10 | 0     | 0          |
| 6         | 10 | 10 | 0     | 0          |
| **Total** |    |    |       | **0.2**    |

Por lo tanto, el valor calculado de  es 0.2.

## Paso 4: Determinar el Valor Crtico o p-valor

Para un nivel de significancia comn de 0.05 y 5 grados de libertad (nmero de categoras - 1 = 6 - 1 = 5), podemos buscar en la tabla de distribucin  el valor crtico correspondiente. Supongamos que este valor es 11.07 (este valor puede variar segn la tabla especfica que ests utilizando).

## Paso 5: Tomar una Decisin

Comparamos el valor calculado de  con el valor crtico de la tabla:

- Si el valor calculado de  (0.2) es menor que el valor crtico (11.07), entonces no hay suficiente evidencia para rechazar la hiptesis nula.
- Por lo tanto, no podemos concluir que el dado no es justo basndonos en nuestros datos.

```python
from scipy.stats import chi2

# Grados de libertad
df = 5
# Nivel de significancia
alpha = 0.05
# Valor observado
chi_square_observed = 0.2

# Valor crtico de chi cuadrado
chi_square_critical_value = chi2.ppf(1 - alpha, df)

# Comparacin del valor observado con el valor crtico
if chi_square_observed > chi_square_critical_value:
    print(f"El valor observado {chi_square_observed} es mayor que el valor crtico {chi_square_critical_value}: Rechazar la hiptesis nula.")
else:
    print(f"El valor observado {chi_square_observed} es menor que el valor crtico {chi_square_critical_value}: No hay suficiente evidencia para rechazar la hiptesis nula.")
```

## Conclusin

Con base en la prueba de bondad de ajuste  realizada, no hemos encontrado evidencia suficiente para rechazar la hiptesis de que el dado es justo. Sin embargo, es importante recordar que "no rechazar H" no es lo mismo que "aceptar H". Simplemente no tenemos suficientes pruebas para decir que el dado no es justo.

## Limitaciones de la Prueba de Bondad de Ajuste 

- Tamao de la muestra: La prueba puede no ser confiable si el tamao de la muestra es muy pequeo.
- Frecuencias bajas: La precisin disminuye cuando las frecuencias esperadas en alguna de las categoras son menores a 5.
- Independencia: La prueba asume que las observaciones son independientes entre s.


```python
import numpy as np
from scipy.stats import chi2

# Definicin de los conjuntos de datos
datos1 = [
    38, 33, 25, 5, 69, 35, 98, 52, 12, 79,
    50, 46, 95, 42, 49, 11, 78, 34, 2, 43,
    7, 50, 5, 91, 77, 18, 21, 4, 17, 62,
    91, 36, 48, 88, 52, 76, 99, 73, 82, 90
]

datos2 = [
    0.8797, 0.3884, 0.6289, 0.8750, 0.5999, 0.8589, 0.9996, 0.2415, 0.3808, 0.9606,
    0.9848, 0.3649, 0.7977, 0.5841, 0.8147, 0.6431, 0.7387, 0.5613, 0.0318, 0.7401,
    0.4557, 0.1592, 0.8536, 0.8846, 0.3410, 0.1492, 0.8681, 0.5291, 0.3188, 0.5592,
    0.9170, 0.2204, 0.5991, 0.5461, 0.5739, 0.3254, 0.0856, 0.2258, 0.4603, 0.5027,
    0.8376, 0.6235, 0.3681, 0.2088, 0.1525, 0.2006, 0.4720, 0.4272, 0.6360, 0.0954
]

# Nmero de intervalos elegidos para segmentar los datos;
num_intervalos1 = 5  # Nmero de categoras para el primer conjunto
num_intervalos2 = 5  # Nmero de categoras para el segundo conjunto

# Clculo de las frecuencias observadas para cada categora de datos basado en los intervalos.
# np.histogram divide los datos en intervalos continuos y cuenta cuntos datos caen en cada intervalo.
frecuencias_observadas1, _ = np.histogram(datos1, bins=np.linspace(min(datos1), max(datos1), num_intervalos1 + 1))
frecuencias_observadas2, _ = np.histogram(datos2, bins=np.linspace(min(datos2), max(datos2), num_intervalos2 + 1))

# Se asume una distribucin uniforme bajo la hiptesis nula para calcular las frecuencias esperadas.
# Cada categora debera tener una cantidad igual de observaciones bajo la hiptesis nula.
frecuencias_esperadas1 = [len(datos1) / num_intervalos1] * num_intervalos1
frecuencias_esperadas2 = [len(datos2) / num_intervalos2] * num_intervalos2

# Clculo del estadstico Chi cuadrado para ambos conjuntos de datos.
# Esto mide la discrepancia entre las frecuencias observadas y esperadas.
chi_cuadrado1 = ((frecuencias_observadas1 - frecuencias_esperadas1) ** 2 / frecuencias_esperadas1).sum()
chi_cuadrado2 = ((frecuencias_observadas2 - frecuencias_esperadas2) ** 2 / frecuencias_esperadas2).sum()

# Clculo del valor crtico del Chi cuadrado para un nivel de significancia del 95% y (nmero de intervalos - 1) grados de libertad.
# Si el estadstico Chi cuadrado es mayor que este valor crtico, rechazamos la hiptesis nula de uniformidad.
valor_critico1 = chi2.ppf(0.95, num_intervalos1 - 1)
valor_critico2 = chi2.ppf(0.95, num_intervalos2 - 1)

# Determinar si los datos observados se desvan significativamente de lo que se esperara bajo una distribucin uniforme.
resultado1 = 'Rechazar' if chi_cuadrado1 > valor_critico1 else 'No rechazar'
resultado2 = 'Rechazar' if chi_cuadrado2 > valor_critico2 else 'No rechazar'

print(f'Resultado para datos1: {resultado1}, Chi cuadrado: {chi_cuadrado1}, Valor crtico: {valor_critico1}')
print(f'Resultado para datos2: {resultado2}, Chi cuadrado: {chi_cuadrado2}, Valor crtico: {valor_critico2}')
```
